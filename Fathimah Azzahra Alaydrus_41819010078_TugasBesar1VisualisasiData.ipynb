```python
tb1 = pd.read_csv ("iris (2).csv", header=None, names=["Panjang Sepal (cm)", "Lebar Sepal (cm)", "Panjang Petal (cm)",
                                                     "Lebar Petal (cm)", "Kelas Atribut"])
```

```python
tb1.head()
```

```python
#min, max, mean, standar deviasi
print (tb1.describe())
```

```python
#Korelasi
dtkorel=tb1.corr()
dtkorel
```

```python
#bar chart
sns.barplot(x="Kelas Atribut", y="Panjang Sepal (cm)",  data=tb1)
plt.title('Bar Chart')
plt.show()
```

```python
#histogram
plt.hist(x="Panjang Sepal (cm)", data=tb1)
plt.show
```

```python
#grafik garis
sns.lineplot(data=tb1)
plt.title('Iris Dataset')
plt.show()
```

```python
#scatter plot 
plt.scatter(x="Panjang Sepal (cm)",y="Lebar Sepal (cm)",data =tb1)
plt.show()
```

```python
#scatter plot
sns.FacetGrid(tb1, hue="Kelas Atribut", height=6) \
   .map(plt.scatter, "Panjang Sepal (cm)", "Lebar Sepal (cm)") \
   .add_legend()
```

```python
#scatter plot
sns.scatterplot(x="Panjang Sepal (cm)",y="Lebar Sepal (cm)",data =tb1, hue="Kelas Atribut")
plt.show()
```
